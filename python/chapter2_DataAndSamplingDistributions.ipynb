```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
loan_inc = np.genfromtxt("loans_income.csv", delimiter=",")
loan_inc
```

```python
sample_5 = [[x for x in np.random.choice(a=loan_inc, size=5)] for y in range(1000)]
sample_mean_5 = [np.mean(x) for x in sample_5]

sample_10 = [[x for x in np.random.choice(a=loan_inc, size=10)] for y in range(1000)]
sample_mean_10 = [np.mean(x) for x in sample_10]

sample_20 = [[x for x in np.random.choice(a=loan_inc, size=20)] for y in range(1000)]
sample_mean_20 = [np.mean(x) for x in sample_20]
```

```python
fig, ax = plt.subplots(3, 1, figsize=(12, 10))
ax[0].hist(sample_mean_5)
ax[1].hist(sample_mean_10)
ax[2].hist(sample_mean_20)
```
